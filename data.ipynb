```python
import pandas as pd

player_df = pd.read_csv("player_data.csv")
player_df.head()
```

```python
player_df.isna().sum()
```

```python
player_df.info()
```

```python
# entferne spalten mit null wert
player_df.dropna(axis="columns", inplace=True, how="all")
player_df.isna().sum()
```

```python
player_df.dropna(inplace=True, how="all")
player_df.isna().sum()
```

```python
# min Spalten angeben die Werte beinhalten müssen 
# zb in 14 Spalten müssen 12 mit Werten befüllt sein
player_df.dropna(inplace=True, thresh=12)
player_df.isna().sum()
```

```python
player_df.reset_index(inplace=True, drop=True)
player_df.info()
```

```python
player_df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Erstelle eine Liste mit allen Spaltennamen ohne die Id
cols = list(player_df.iloc[:, 1:])

# 3x5 Matrix
fig, axes = plt.subplots(3, 5, figsize=(18, 11))

# Abstand zwischen den Diagrammen
fig.tight_layout(pad=2.0)

# for Schleife über cols laufen lassen
for i in range(len(cols)):
    sns.boxplot(ax=axes[i // 5, i % 5], y=player_df[cols[i]])
```

```python
# minimum Wert bei points herausfinden
points_outlier = player_df["points"].idxmin()
points_outlier
```

```python
possession_outlier = player_df["possessions"].idxmin()
possession_outlier
player_df.drop(player_df.index[points_outlier], inplace=True)
player_df.tail(10)
player_df.reset_index(inplace=True, drop=True)
player_df.info()
```
